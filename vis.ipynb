```python
import obspy
import matplotlib.pyplot as plt
```

```python
rf = obspy.read('./build/src/test_RF.sac')[0]
rf_ref = obspy.read('./test_RF_ref.sac')[0]
time_cpu = 3.07400703
time_gpu = 0.158917516
shift = 10
plt.figure(figsize=(6, 6))
plt.subplot(2, 1, 1)
plt.plot(rf.times()-shift, rf.data, label='RF (time = 0.227114931s on GPU) ', lw=4)
plt.plot(rf_ref.times()-shift, rf_ref.data, label='RF (time = 3.07400703s on CPU) ', linestyle='dashed')
plt.legend()
plt.xlim(-5, 30)
plt.xlabel('Time (s)')

plt.subplot(2, 1, 2)
# plot time comparison
plt.barh(['GPU', 'CPU'], [time_gpu, time_cpu], color=['C0', 'C1'])
plt.xlabel('Calculation Time (s)')
# anotate speed up
plt.text(time_gpu + 0.05, 0, f'{time_cpu / time_gpu:.1f}x speed up', va='center')
plt.tight_layout()
```
